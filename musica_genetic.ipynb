```python
from music21 import *
import random
import numpy as np
```

```python
scala = ['C4','C#4','D4','D#4','E4','F4','F#4','G4','G#4','A4','A#4','B4']

def genera_lista():
    lista = []
    for i in range(0, 16):
        lista.append(random.randint(0,11))
    return lista

def note_a_numeri(lista, scala):
    lista = lista.copy()
    scala = scala.copy()
    for i in range(0, 16):
        lista[i] = scala.index(lista[i])
    return lista
        
def esporta(sequenza, nome_file):
    scala = ['C4','C#4','D','D#4','E','F','F#','G','G#','A','A#','B']
    stream1 = stream.Stream()
    for nota in sequenza:
        stream1.append(note.Note(scala[nota]))
    stream1.write('midi', fp='T://datasets/musica//'+nome_file)

    
    
def crossover(lista1, lista2, probabilità):
    lista1 = lista1.copy()
    lista2 = lista2.copy()
    for i in range(0, 15):
        ran = random.randint(1, 100)
        if ran < probabilità:
            lista2[i], lista1[i] = lista1[i], lista2[i]
    return lista1, lista2

def mutazione(lista, probabilità): #5%
    lista = lista.copy()
    for i in range(0, 15):
        ran = random.randint(1, 100)
        segno = random.choice([True, False])
        if ran < probabilità:
            if i == 0:
                if segno:
                    if lista[0] == 15:
                        lista[0] -= 1
                    else:
                        lista[0] += 1
                if not segno:
                    lista[0]
            if i != 0:
                n = lista[i]
                sc = scale.MajorScale('c4')
                m = lista[i-1]
                n = scala[n]
 
                if segno:
                    m = sc.next(n, 'descending', stepSize=2)
                    m = str(m)
                if not segno:
                    m = sc.next(n, 'ascending', stepSize=2)
                    m = str(m)
                if m in scala:
                    valore = scala.index(m)
                    lista[i] = valore
                
    return lista
                
    

def genitori(popolazione, fitness):
    def getKey(item):
        return item[0]
    def genitore(popolazione, fitness):
        popolazione = popolazione.copy()
        fitness = fitness.copy()
        
        ran_int = []
        for i in range(0,3):
            ran_int.append(random.randint(0, 19))
        print('seq = ', str(ran_int))
        tupla = []
        for i in ran_int:
            tupla.append([fitness[i], i])
        sort = sorted(tupla, key=getKey)
        vincitore = sort[2]
        try:
            genitore = popolazione[vincitore[2]]
        except:
            try:
                genitore = popolazione[vincitore[1]]
            except:
                genitore = popolazione[vincitore[0]]
                               
                
        print(genitore)
        return genitore
        
    gen1 = genitore(popolazione, fitness)
    gen2 = genitore(popolazione, fitness)
    return gen1, gen2
            
            
            
        
        
def elitarismo(popolazione1, popolazione2, fitness1, fitness2):
    fitness1 = fitness1.copy()
    fitness2 = fitness2.copy()
    popolazione1 = popolazione1.copy()
    popolazione2 = popolazione2.copy()
    peggiore = min(fitness2)
    migliore = max(fitness1)
    indice_peggiore = fitness2.index(peggiore)
    indice_migliore = fitness1.index(migliore)
    del popolazione2[indice_peggiore]
    del fitness2[indice_peggiore]
    sequenza_migliore = popolazione1[indice_migliore]
    popolazione2.append(sequenza_migliore)
    print('a', str(len(popolazione1)))
    fitness2.append(migliore)
    return popolazione2, fitness2
```
